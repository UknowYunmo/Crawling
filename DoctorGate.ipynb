```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import urllib.request
import time
#f=open('c:\\data\\gimo.txt','w',encoding='UTF-16')
binary = 'C:/Users/NICE/Downloads/chromedriver_win32/chromedriver.exe'
browser = webdriver.Chrome(binary)
browser.get("http://data.medigate.net/location")
browser.maximize_window()
time.sleep(2)
browser.implicitly_wait(2)
html = browser.page_source
soup = BeautifulSoup(html, "lxml")
browser.find_element_by_class_name('jss5').click()
browser.find_element_by_class_name('MuiButton-label').click()
time.sleep(2)
browser.implicitly_wait(2)
browser.find_element_by_xpath('//*[@id="root"]/div[2]/div[1]/div/div/div[1]/div[2]/div').click()
browser.find_element_by_xpath('//*[@id="popperSpecialty"]/div/div[3]/div[1]/div').click()
browser.find_element_by_xpath('//*[@id="popperSpecialty"]/div/div[1]/div[2]/button/span[1]').click()
browser.find_element_by_xpath('//*[@id="map"]/button').click()
time.sleep(2)
browser.implicitly_wait(2)
browser.find_element_by_xpath('/html/body/div[2]/div[3]/div/div/div[1]/div[3]/button').click()
time.sleep(2)
browser.implicitly_wait(2)
html = browser.page_source
soup = BeautifulSoup(html, "lxml")
soup.find_all('div', class_ ='aria-label')
```
